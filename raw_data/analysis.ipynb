```python
import pandas as pd
```

```python
df = pd.read_csv('LoanApp.csv.gz')
df
```

```python
def compute_cycle_time(df, start_timestamp_col='start_timestamp', end_timestamp_col='end_timestamp', case_id_col='case_id'):
    df[start_timestamp_col] = pd.to_datetime(df[start_timestamp_col], format='mixed')
    df[end_timestamp_col] = pd.to_datetime(df[end_timestamp_col], format='mixed')

    # Calculate duration per case
    df['case_duration'] = df.groupby(case_id_col).apply(lambda x: (x[end_timestamp_col].max() - x[start_timestamp_col].min()).total_seconds() / 3600).reset_index(drop=True)

    # Compute average case duration
    average_duration = df['case_duration'].mean()

    return average_duration
```

```python
average_duration = compute_cycle_time(df, start_timestamp_col='start_time', end_timestamp_col='end_time', case_id_col='case_id')
average_duration
```
